```python
import pandas as pd
import numpy as np
import yfinance as yf
import warnings
warnings.filterwarnings("ignore",category=FutureWarning)
import requests
import json
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Download historical data for a stock using yfinance
tesla = yf.Ticker('TSLA')
tesla_data = tesla.history(period='max')
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
#Webscraping to Extract Tesla Revenue Data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")
```

```python
tesla_revenue_list = []

tables = soup.find_all("table")

for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        for row in table.tbody.find_all("tr"):
            cols = row.find_all("td")
            if len(cols) == 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip()
                tesla_revenue_list.append({"Date": date, "Revenue": revenue})
        break
tesla_revenue = pd.DataFrame(tesla_revenue_list)
```

```python
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"",regex=True)
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
tesla_revenue.tail()
tesla_data.to_csv("tesla_stock.csv", index=False)
tesla_revenue.to_csv("tesla_revenue.csv", index=False)
```

```python
make_graph(tesla_data,tesla_revenue,"Tesla")
```

```python
#Use yfinance to Extract Stock Data
gamestop = yf.Ticker('GME')
gme_data=gamestop.history(period='max')
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
#Webscraping to Extract GME Revenue Data
url2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
html_data_2 = requests.get(url2).text
soup2 = BeautifulSoup(html_data_2, "html.parser")

gme_revenue_list = []

tables = soup2.find_all("table")

for table in tables:
    if "GameStop Quarterly Revenue" in table.text:
        for row in table.tbody.find_all("tr"):
            cols = row.find_all("td")
            if len(cols) == 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip().replace(",", "").replace("$", "")
                gme_revenue_list.append({"Date": date, "Revenue": revenue})
        break


gme_revenue = pd.DataFrame(gme_revenue_list)
gme_data.to_csv("gme_stock.csv", index=False)
gme_revenue.to_csv("gme_revenue.csv", index=False)
gme_revenue.tail(5)
```

```python
make_graph(gme_data, gme_revenue, "GameStop")
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import pandas as pd

def make_graph(stock_data, revenue_data, company_name):
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])
    stock_data.sort_values('Date', inplace=True)

    revenue_data['Revenue'] = revenue_data['Revenue'].replace({'\$': '', ',': ''}, regex=True)
    revenue_data['Revenue'] = pd.to_numeric(revenue_data['Revenue'], errors='coerce')
    revenue_data['Date'] = pd.to_datetime(revenue_data['Date'])
    revenue_data.dropna(subset=['Revenue'], inplace=True)
    revenue_data.sort_values('Date', inplace=True)

    #subplots
    fig = make_subplots(
        rows=2, cols=1,
        shared_xaxes=True,
        subplot_titles=("Historical Share Price", "Revenue"),
        vertical_spacing=0.2
    )

    #Stock price trace
    fig.add_trace(go.Scatter(
        x=stock_data['Date'],
        y=stock_data['Close'],
        name="Stock Price",
        mode='lines',
        line=dict(color='blue')
    ), row=1, col=1)

    #Revenue trace
    fig.add_trace(go.Scatter(
        x=revenue_data['Date'],
        y=revenue_data['Revenue'],
        name="Revenue",
        mode='lines',
        line=dict(color='blue')
    ), row=2, col=1)

    #Layout
    fig.update_layout(
        title=company_name,
        height=700,
        template="plotly_white"
    )

    fig.update_yaxes(title_text="Stock Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US)", row=2, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)

    fig.show()
```
